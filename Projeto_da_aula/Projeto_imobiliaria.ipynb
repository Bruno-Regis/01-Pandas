## Análise Exploratória de Dados

#### Qual o valor médio de aluguel por tipo de imóvel?

Média de todos imóveis da coluna Valor:

```python
dados['Valor'].mean()
```

Agrupando a coluna Tipo de imóvel às médias de todos valores numéricos apenas

```python
dados.groupby('Tipo').mean(numeric_only = 'True')
```

Agrupando em função apenas dos Preços dos imóveis, descritos na coluna Valor

```python
dados.groupby('Tipo')['Valor'].mean()
```

Melhorando a visualização, transformando Valor em um dataframe adicionando [], além do mais, utilizei sort_values para classificar os valores em ordem crescente

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Plotando gráfico de barras horizontais com os valores agrupados das médias dos imóveis

```python
df_tipo_valor = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_valor.plot(kind='barh', figsize=(10,8), color = 'green')
```

#### Removendo Imóveis Comerciais

Tipos de imóveis:

```python
dados.Tipo.unique()
```

Lista de imóveis a serem removidos:

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

Utilizando a função query, a condição que passamos para o método query() é sempre no formato de uma expressão, entre aspas.
Abaixo, seleciono todos as linhas do dataframe que o tipo de imóvel esteja dentro da lista imóveis_comerciais

```python
dados.query('@imoveis_comerciais in Tipo')
```

No caso, queremos o contrário, todos os imóveis que não sejam comerciais, para tal utilizo a expressão not in:

```python
dados.query('@imoveis_comerciais not in Tipo')
```

Guardando em uma variável = df_imoveis_residenciais

```python
df_imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')
```

```python
df_imoveis_residenciais.head()
```

Verificando se todos os tipos presentes no df_imoveis_residenciais realmente não contem imóveis comerciais:

```python
df_imoveis_residenciais.Tipo.unique()
```

Plotando novamente o gráfico de barras horizontais com Df de imóveis residenciais

```python
df_tipo_valor = df_imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_valor.plot(kind='barh', figsize=(10,8), color = 'green')
```

#### Percentual de cada tipo de imóvel presente na base de dados

```python
#contando valores a série Tipo do dataframe e normalizando para aparecer em percentual, além de utilizar o método to frame
print(df_imoveis_residenciais.Tipo.value_counts(normalize= 'True',).sort_values().to_frame(name = 'Percentual'))
percentuais_residenciais = df_imoveis_residenciais.Tipo.value_counts(normalize= 'True',).sort_values().to_frame(name = 'Percentual')
```

Plotando gráfico de barras para visualização dos  resultados:

```python
percentuais_residenciais.plot(kind='bar', figsize=(12,8), color = 'green',edgecolor='black', 
                              xlabel= 'Tipo de imóvel', ylabel= 'Percentual de imóveis')
```

Os imóveis do tipo apartamento compreendem aproximadamente 84% do total de imóveis residenciais.
